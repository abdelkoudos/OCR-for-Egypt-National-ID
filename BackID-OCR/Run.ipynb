# EVALUATION FOR SINGLE PATH

```python
import glob
import os
from ID_Recognizer_py import BackID_Recognizer

# --- CONFIG & RUN ---
config = {
    'resize_width': 760,
    'resize_height': 32, # alt config opt: 36
    'blockSize': 75,
    'C': 20,
    'iou_threshold': 0.0001,
    'box_width': 35 # alt config opt: 36
}

model_path = "char_recognizer_model_f3.h5" # alt model char_recognizer_model_f2.h5
image_path = "test.jpg"


# Create Object
recognizer = BackID_Recognizer(config=config, model_path=model_path, image_path=image_path)
```

```python
result = recognizer.predict_image()
print(f"[Predicted ID]: {''.join(map(str, result))}\n")
```

```python
image_dir = "Images/"
image_paths = glob.glob(os.path.join(image_dir, '*.jpg'))
for path in image_paths[:]:
    recognizer = BackID_Recognizer(config=config, model_path=model_path, image_path=path)
    
    try:
        result = recognizer.predict_image()
        print(f"[image pth ID]: {path[-18:-4]}")
        print(f"[Predicted ID]: {''.join(map(str, result))}\n")
        

    except:
        print(f"[Warning]: Something went wrong at {path[-18:-4]} - Try change config! \n")

```
